```python
from transformers import pipeline
pipe = pipeline("text-to-speech", model="suno/bark-small",device="cuda")
text = "Okay, so here we're defining a function `dfs` that takes in a graph and a starting node. The graph is represented as a dictionary where each key is a node and the value is a list of its neighbors. The function will return a list of nodes in the order they were visited."
output = pipe(text)
print(output["audio"].shape)
print(output["sampling_rate"])
from IPython.display import Audio
Audio(output["audio"], rate=output["sampling_rate"])
```
